```python
import pandas as pd
import numpy as np

from sklearn.calibration import CalibratedClassifierCV, calibration_curve, CalibrationDisplay
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier
from sklearn.linear_model import LogisticRegression
import xgboost as xgb

from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
import joblib

from sklearn.metrics import classification_report
```

```python
Train = pd.read_csv('data\Train_Data.csv').drop('id', axis='columns')
test = pd.read_csv('data\Test_Data.csv').drop('id', axis='columns')
```

```python
Train
```

```python
x_train, x_val, y_train, y_val = train_test_split(Train.drop('default', axis='columns'),
                                                  Train.default, train_size=0.60)
```

```python
x_test = test.drop('default', axis='columns')
y_test = test.default
```

```python
model_rf = RandomForestClassifier().fit(x_train, y_train)
model_xgb = xgb.XGBClassifier(n_estimators=1000,
                              early_stopping_rounds=200).fit(x_train, y_train,
                                                            eval_set=[(x_val,y_val)],
                                                            verbose=False)
```

```python
joblib.dump(model_xgb,"models/XGBoostModel.pkl")
joblib.dump(model_rf,"models/RForestModel.pkl")
```
